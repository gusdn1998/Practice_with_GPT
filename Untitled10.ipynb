```python
# By using PCA, we performed dimensionality reduction in a way that reduces the large number of independent variables while preserving as much of their characteristics as possible.
# Instead of selecting a subset of the original variables, we generated new variables that reflect all of the information from the original ones.
# In the second task, we also compared the model training time before and after applying PCA to see how much it could be reduced.
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/snkn1959/data_source/main/customer_pca.csv')
df.head()
```

```python
X = df.drop('label', axis = 1)
y = df['label']
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
# Here, n_components = 2 means that the original data will be reduced to two dimensions. In other words, two principal components (PC1 and PC2) will be created.
X = pd.DataFrame(pca.fit_transform(X), columns = ['PC1','PC2'])
# fit_transform() finds the principal components based on the data and then projects the data onto these new principal component axes.
# The result is the reduced 2-dimensional data (PC1 and PC2), which is then converted into a pandas DataFrame with the variable names set to PC1 and PC2.
X.head()
#This code reduces the original data X to two principal components, replaces it with the new variables PC1 and PC2, and displays a portion of the transformed data.
```

```python
X = X.join(y)
X.head()
```

```python
sns.scatterplot(x = 'PC1', y = 'PC2', data = X, hue = 'label',\
               palette = 'rainbow')
plt.show()
```

```python
print(df.shape)
print(df.drop('label', axis=1).shape)
print(df.drop('label', axis=1).columns)
print(pca.components_.shape)
```

```python
from sklearn.decomposition import PCA
import pandas as pd

X = df.drop('label', axis=1)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

df_comp = pd.DataFrame(
    pca.components_.T,
    columns=['PC1', 'PC2'],
    index=X.columns
)

print(df_comp)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(
    df_comp,
    annot=True,
    cmap='coolwarm',
    center=0
)
plt.title('PCA Component Loadings Heatmap')
plt.xlabel('Principal Components')
plt.ylabel('Original Variables')
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.show()
```

```python
# Intuitively, index 1 (PC2) shows a relatively strong negative correlation with the gas_transport variable and a relatively strong positive correlation with the travel variable.
# This means that the larger the PC2 component, the lower the value associated with gas_transport and the higher the value associated with travel.
```

```python
df = pd.read_csv('https://media.githubusercontent.com/media/musthave-ML10/data_source/main/anonymous.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split

xtr, xt, ytr, yt = train_test_split(df.drop('class', axis = 1), df['class'], test_size = 0.2,\
                                   random_state = 100)

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
scaled_xtr = sc.fit_transform(xtr)

scaled_xt = sc.transform(xt)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state = 100)


import time
start_time = time.time()

rf.fit(scaled_xtr, ytr)

print(time.time() - start_time)
```

```python
# This workflow uses a high-dimensional dataset with over 4,000 features to train a Random Forest classifier.
# The data is split into training and test sets, then standardized using StandardScaler to normalize feature scales.
# The model is trained on the scaled training data, and the training time is measured.
# This setup can be used to compare model performance and training time before and after dimensionality reduction, such as with PCA.
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score

predrf = rf.predict(scaled_xt)
accuracy_score(yt, predrf)
```

```python
probarf = rf.predict_proba(scaled_xt)
roc_auc_score(yt, probarf[:,1])
```

```python
# Before applying PCA, the modeling took about 93 seconds, and both the accuracy score and ROC AUC score were very high.
```

```python
pca= PCA(n_components = 2)

pca.fit(scaled_xtr)
```

```python
pca.explained_variance_ratio_
```

```python
# Combined, these two components account for only about 0.08, meaning they reflect roughly 8% of the information in the original data.
# While the optimal number of principal components can be somewhat subjective, I attempted to apply the elbow method—used earlier for the K-means clustering—to help determine an appropriate cutoff.
```

```python
var_ratio = []

for i in range(100, 550, 50):

    pca = PCA(n_components = i)

    pca.fit_transform(scaled_xtr)

    ratio = pca.explained_variance_ratio_.sum()

    var_ratio.append(ratio)
```

```python
sns.lineplot(x = range(100,550,50), y = var_ratio)
```

```python
pca = PCA(n_components = 400, random_state = 100)

pca_xtr = pca.fit_transform(scaled_xtr)

pca_xt = pca.transform(scaled_xt)
```

```python
rf2 = RandomForestClassifier(random_state = 100)

start_time = time.time()
rf2.fit(pca_xtr, ytr)
print(time.time() - start_time)
```

```python
# After applying PCA, the Random Forest modeling took about 57 seconds, showing a significant improvement in training time compared to before.
```

```python
pred2 = rf2.predict(pca_xt)
accuracy_score(yt, pred2)
```

```python
proba2 = rf2.predict_proba(pca_xt)
roc_auc_score(yt, proba2[:,1])
```
